### Trend Analysis

```python
# Cacuating Moving Average

Asian_market['MA_50']= Asian_market['CloseUSD'].rolling(window= 50).mean()

```

```python
# Determine the trend based on moving average
Asian_market['trend'] = np.where(Asian_market['CloseUSD'] > Asian_market['Movement'], 'Up', 'Down')
```

```python
plt.figure(figsize=(10,5))
plt.plot(Asian_market['Date'], Asian_market['CloseUSD'], label = 'Close Price')
plt.plot(Asian_market['Date'], Asian_market['MA_50'], label= '50-day Moving Average')
plt.title('Closing Price and 50-day Moving Average')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```
